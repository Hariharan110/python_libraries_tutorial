# **Histogram :**
  > histogram shows `frequency` distribution.
  It is a graph showing the number of observations within each given interval
  1. histtype = type of graph

```python
marks= [76,45,34,13,25,98,56,87,23,87]
grade_intervels = [0,35,70,100]
```

```python
plt.title("student grade")
plt.hist(marks,grade_intervels)
```

```python
plt.title("student grade")
plt.hist(marks,grade_intervels)
plt.xlabel("percentage")
plt.ylabel("no.of students")
```

```python
# histtype = 'bar','step',stepfield,
# facecolor= k,r,
plt.subplot(2,2,1)
plt.hist(marks,grade_intervels,histtype="bar",facecolor='r')

plt.subplot(2,2,2)
plt.hist(marks,grade_intervels,histtype="step",facecolor='b')

plt.subplot(2,2,3)
plt.hist(marks,grade_intervels,histtype='barstacked')

plt.subplot(2,2,4)
plt.hist(marks,grade_intervels,histtype="stepfilled",facecolor="b")
```

```python
# rwidth=0.5
plt.hist(marks,grade_intervels,histtype="bar",rwidth=0.6)
```

# **Methods ():**
 1. xlabel(),ylabel()  = naming the axis values
 2. xticks(),yticks()  = it change the value of x labels
 3. xlim(),ylim() = it arange the value x and y axis

```python
 x1=['hari','deva','christi','ajay','siddu']
 y1=[76,34,67,89,90]
```

```python
# xlabel(),ylabel()

plt.plot(x1,y1)
plt.title("labels")
plt.xlabel("Name")
plt.ylabel("marks")
```

```python
# xticks(),yticks()

plt.plot(x1,y1)
plt.title("ticks()")
plt.xticks(x1,["a","b","c","d","e"])
plt.yticks(y1,["1","2","3","4","5"])
plt.show()
```

```python
#xlim(),ylim()
plt.plot(x1,y1)
plt.title("ticks()")
plt.xlim([1,5])
plt.ylim([10,100])
```

```python
x=[1,3,5,7,9]
plt.plot(x,y1)
plt.title("ticks()")
plt.xlim([1,10])
plt.ylim([10,100])
```
