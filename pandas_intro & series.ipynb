**Series:**
* we can ceate series in 4 ways
  1. empty Series
  2. Series with array
  3. Series with list
  4. Series with dictionary


```python
import pandas as pd
```

```python
empty=pd.Series()
empty
```

```python
#Series with array
import numpy as np
arr=np.array([1,2,3,4,5])
arr_series=pd.Series(arr)
arr_series
```

```python
# series with list
lst=[3,4,5,7,8]
lst_series=pd.Series(lst,index=[1,2,3,4,5])
lst_series
```

```python
# series with dictionary

dic={"name":"hari","age":20,"salary":10000}
print(dic)
print("-------------")
dic_series=pd.Series(dic)
print(dic_series)
print("-------------")
for i in dic_series.items(): #dic_series.items() this return both key: value
  print(i)
```

```python
pd.Series(dic,index=["name","age"]) # create series olny with 2 values
```

**DataFrame**
* wasy to create dataframe
  1. load data - dictionary or list
  2. load data - excel or csv
  3. load data - list of tuple

```python
#list dataFrame
lst=[['hari','david','mani'],[21,432,43],['ooty','madurai','svg']]
df=pd.DataFrame(lst)
print(df)
print('----------------------')
lst1 = [['hari',21,'svg'],['deva',34,'kovai'],['pirai',20,'chennai']]
df1=pd.DataFrame(lst1,columns=['name','age','city'])
print(df1)
```

```python
#dictionary dataframe
data={"name":{1:"hari",2:"david",3:"mani",4:"vijay",5:"mathavan"},
      "age":{1:21,2:24,3:32,4:43,5:12},
      "salary":{1:2000,2:4000,3:1000,4:2300,5:5500}}

data_se=pd.Series(data)       # series
data_df=pd.DataFrame(data)    # dataframe
print(data_se)
print(data_df)
```

```python
data={"name":{"hari","david","mani","vijay","mathavan"},
      "age":{21,24,32,43,12},
      "salary":{2000,4000,1000,2300,5500}}                 #erroe bcz in dictionary there is a set{} value its un orderd , if i change it, it will run
df=pd.DataFrame(data)
df
```

```python
data={"name":["hari","david","mani","vijay","mathavan"],
      "age":[21,24,32,43,12],
      "salary":[2000,4000,1000,2300,5500]}                 #it will run bcz inside the dic there is list value its ordered it wiil run
df=pd.DataFrame(data)
df
```

```python
#load data list of tuples
lst=[("hari",21),("vijar",20),("matha",31)]
lst1=pd.DataFrame(lst)
lst1
```
