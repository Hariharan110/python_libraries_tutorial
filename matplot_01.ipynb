```python
a=[20,30,12,57]
b=[87,34,56,13]
x=np.arange(1,50,13)
y=np.random.randint(1,90,4)
```

# **Suplot();**
1. suplot(R,C,I)
   >> R =row ,
   >> C = column ,
   >> I =  index

```python
x=[1,2,3,4,5]
y1=[6,4,8,9,1]
y2=np.random.randint(1,10,5)
y3=np.random.randint(1,10,5)
y4= np.random.randint(1,10,5)
```

```python
# creating subplot-1
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('(a)')

# creating subplot-2
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('(b)')

# creating suplot -3
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('(c)')

# creating subplot-3
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('(d)')

plt.suptitle("suplots")  # for subplots giving title using -- suptitle()
```

# **legend();**
1. **loc()** = where to locate the labels   ---('best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
)
2. **framealpha()** = Transperancy of label box ( 0 = invisible , 1 = visible)
3. **facecolor()** =  lable box inside color
4. **edgecolor()** = label box outline color
5. **fancybox()** = shape of label( True = rectangle , false=rounded rectangle)
6. **shadow()** = shadow of the label (true,false)

```python
x=[1,2,3,4,5]
y1=[6,4,8,9,1]
y2=np.random.randint(1,10,5)
# y3=np.random.randint(1,10,5)
# y4= np.random.randint(1,10,5)
```

```python
plt.plot(x,y1,label='age')
plt.plot(x,y2,label='number')
# this will not show the labels ,bcz legend() didn't used
```

```python
plt.plot(x,y1,label='age')
plt.plot(x,y2,label='number')
plt.legend()
```

```python
from matplotlib.patches import Shadow
plt.plot(x,y1,x,y2)
plt.legend(['age','no'],loc='upper left', framealpha = 1 , facecolor = 'c' , edgecolor = 'r', fancybox =False , shadow =True)
```
